# レース結果のスクレイピング

```python
import pandas as pd
import time
from tqdm import tqdm # for prgress bar
```

### スクレイピングする年をtarget_yearへ入力

```python
target_year = '2020'
```

### スクレイピング関数を定義

```python
def scrape_race_results(race_id_list, pre_race_results={}):
    race_results = pre_race_results
    for race_id in tqdm(race_id_list):
        if race_id in race_results.keys():
            continue
        try:
            time.sleep(1)
            url = 'https://db.netkeiba.com/race/' + race_id
            race_results[race_id] = pd.read_html(url)[0]
        except IndexError:
            continue
        except Exception as e:
            print(e)
            break
        except:
                break
    return race_results
```

### スクレイピングで使用するレースIDリストを作成

```python
race_id_list = []
for place in range(1,11,1):
    for kai in range(1,6,1):
        for day in range(1,13,1):
            for r in range(1,13,1):
                race_id = target_year + str(place).zfill(2) + str(kai).zfill(2) + str(day).zfill(2) + str(r).zfill(2)
                race_id_list.append(race_id)
```

### スクレイピング開始

```python
rawdata = {}
rawdata = scrape_race_results(race_id_list, rawdata)
```

### 取得データをピックルファイルで保存

```python
for key in rawdata: # key を取り出し
    # .indexに、各レース毎の列数がRangeIndex型で入っているので、
    # それをkey値*列数分代入
    rawdata[key].index = [key] * len(rawdata[key])
    
# for文でkeyを取得し、そのkeyを使ってtest6[key]で1レース分のデータを引っ張ってくる
# [rawdata[0], rawdata[1]...] みたいになるイメージ
results = pd.concat([rawdata[key] for key in rawdata], sort = False)

# ファイルに保存
file_name = target_year + '_result_raw,pickle'
results.to_pickle(file_name)
```
